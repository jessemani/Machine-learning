### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### replot

The one we will use most is relplot(). This is a figure-level function for visualizing statistical relationships using two common approaches: scatter plots and line plots. relplot() combines a FacetGrid with one of two axes-level functions:

### We are using the inbuilt dataset 'TIPS'

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);# The inbuilt kind parameter is scatter i.e kind='scatter'
```

```python
tips.head()
```

### Using the hue parameter.The hue paramater must be a Categorical data

```python
sns.relplot(x="total_bill", y="tip",hue='sex',data=tips);
```

### Using the style parameter.The style parameter must be a categrical data

```python
sns.relplot(x="total_bill", y="tip",hue='sex',style='smoker',data=tips);
```

### using the size parameter.The size must be a numerical category.

```python
sns.relplot(x="total_bill", y="tip",hue='sex',style='smoker',size='size',sizes=(15, 200),data=tips);
```

## Categorical Scatterplots
### x-axis is categorical
### y-axis is numerical

```python
sns.catplot(x='day',y='tip',kind='strip',data=tips)# we cam apply 'jitter'parameter to adjust the graph.
```

A strip is basically a scatter plot where the x axis represents a categorical variable. Typical uses of a strip plot involves applying a small random jitter value to each data point such that the separation between points becomes clearer.

```python
sns.catplot(x='day',y='tip',kind='swarm',data=tips) # It provides distrubution of  the plot
```

```python
### We can add hue paramater
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='tip',data=tips,hue='sex')
```

```python
sns.boxplot(tips['tip'])
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box')
```

## vilon plots
vilon plots are considered as boxplot with kde.It is symentrical on both sides and it also shows the distubution of data.These are used to compare the data distrubation across the different categories.

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

## BAR PLOT
A graph drawn using rectangular bars to show how large each value IS.

```python
sns.barplot(x='day',y='total_bill',data=tips)# The black line in that is the confidence intreval estimate of the mean.we can change that by estimator parameter
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.median)
```

```python
sns.countplot(x='sex',data=tips) #barplot is a figure level and countplot is a axis level plot.
```

## DISTPLOT

1. Distplots are used to univariate analysis on continious/numerical data
2. The three types of distplots are:
    1. HISTOGRAM
    2. KDE(KERNEL DENSITY ESTIMATION) OR PDF
    3. RUG PLOT 

```python
tit=sns.load_dataset('titanic')
```

```python
tit.head(10)
```

```python
tit.info()
```

```python
tit.sample()
```

```python
tit.dropna(inplace=True)
```

```python
tit.info()
```

```python
sns.distplot(tit['fare'])
```

## PLOTING A HISTOGRAM

```python
sns.distplot(tit['fare'],bins=90,kde=False)
```

# Ploting only kde

```python
sns.distplot(tit['fare'],bins=90,kde=True,hist=False)
```

# Plotting only RUG plot.Its like a histogram in which points are plotted side by side rather than top.This providers extra information how the points are represented bin.

```python
sns.distplot(tit['fare'],bins=90,kde=False,hist=False,rug=True)
```
